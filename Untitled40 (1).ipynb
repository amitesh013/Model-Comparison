```python
import timm

class GhostNetV2Backbone(nn.Module):
    """
    GhostNetV2 backbone via timm.
    Extracts multi-scale feature maps at strides 8, 16, 32
    (P3, P4, P5) — the standard YOLO neck input format.
    """
    def __init__(self, pretrained=True):
        super().__init__()
        self.model = timm.create_model(
            'ghostnetv2_100',
            pretrained=pretrained,
            features_only=True,          # returns feature pyramid
            out_indices=(2, 3, 4),       # strides 8, 16, 32
        )
        self.out_channels = self.model.feature_info.channels()  # e.g. [40, 80, 160]

    def forward(self, x):
        return self.model(x)   # returns [P3, P4, P5]

# Quick sanity check
_bb = GhostNetV2Backbone(pretrained=False)
_dummy = torch.zeros(1, 3, 640, 640)
_feats = _bb(_dummy)
print("Opt-1 GhostNetV2 output channels:", [f.shape for f in _feats])
del _bb, _dummy, _feats
```

```python
class BiFPNNode(nn.Module):
    """Single BiFPN weighted-fusion node."""
    def __init__(self, channels, num_inputs=2, epsilon=1e-4):
        super().__init__()
        self.epsilon = epsilon
        self.weights = nn.Parameter(torch.ones(num_inputs, dtype=torch.float32))
        self.conv = nn.Sequential(
            nn.Conv2d(channels, channels, 3, padding=1, groups=channels, bias=False),
            nn.Conv2d(channels, channels, 1, bias=False),
            nn.BatchNorm2d(channels),
            nn.SiLU(inplace=True),
        )

    def forward(self, inputs):
        w = F.relu(self.weights)
        w = w / (w.sum() + self.epsilon)
        fused = sum(wi * fi for wi, fi in zip(w, inputs))
        return self.conv(fused)


class BiFPNLayer(nn.Module):
    """
    One BiFPN layer: top-down pass then bottom-up pass.
    Input: [P3, P4, P5]  (from backbone)
    Output: [P3', P4', P5']  (enriched features)
    """
    def __init__(self, channels):
        super().__init__()
        # Top-down
        self.td_p4 = BiFPNNode(channels, num_inputs=2)
        self.td_p3 = BiFPNNode(channels, num_inputs=2)
        # Bottom-up
        self.bu_p4 = BiFPNNode(channels, num_inputs=3)
        self.bu_p5 = BiFPNNode(channels, num_inputs=2)

    def _resize(self, x, target):
        return F.interpolate(x, size=target.shape[2:], mode='nearest')

    def forward(self, feats):
        p3, p4, p5 = feats

        # ── Top-down ──────────────────────────
        p4_td = self.td_p4([p4, self._resize(p5, p4)])
        p3_out = self.td_p3([p3, self._resize(p4_td, p3)])

        # ── Bottom-up ─────────────────────────
        p4_out = self.bu_p4([p4, p4_td,
                              F.avg_pool2d(p3_out, kernel_size=2, stride=2)])
        p5_out = self.bu_p5([p5,
                              F.avg_pool2d(p4_out, kernel_size=2, stride=2)])

        return [p3_out, p4_out, p5_out]


class BiFPNNeck(nn.Module):
    """
    Stacked BiFPN (3 layers).
    Projects backbone channels → uniform BiFPN channels first.
    """
    def __init__(self, in_channels, bifpn_channels=128, num_layers=3):
        super().__init__()
        # Channel unification convs
        self.projs = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(c, bifpn_channels, 1, bias=False),
                nn.BatchNorm2d(bifpn_channels),
                nn.SiLU(inplace=True),
            )
            for c in in_channels
        ])
        self.layers = nn.ModuleList([
            BiFPNLayer(bifpn_channels) for _ in range(num_layers)
        ])
        self.out_channels = [bifpn_channels] * 3

    def forward(self, feats):
        feats = [proj(f) for proj, f in zip(self.projs, feats)]
        for layer in self.layers:
            feats = layer(feats)
        return feats

# Sanity check
_bb = GhostNetV2Backbone(pretrained=False)
_neck = BiFPNNeck(_bb.out_channels, bifpn_channels=128, num_layers=3)
_dummy = torch.zeros(1, 3, 640, 640)
_out = _neck(_bb(_dummy))
print("Opt-2 BiFPN output shapes:", [f.shape for f in _out])
del _bb, _neck, _dummy, _out
```

```python
class CircularPositionalBias(nn.Module):
    """
    Generates a learnable radial distance map centred at the
    feature-map centre. Forces the attention to be aware of
    circular / concentric ring structures — novel for target detection.
    """
    def __init__(self, channels, grid_size=20):
        super().__init__()
        self.grid_size = grid_size
        # Learnable radial sigma (controls how tight the bias is)
        self.sigma = nn.Parameter(torch.tensor(0.5))
        # Small conv to mix positional bias into channel features
        self.mix = nn.Conv2d(channels + 1, channels, 1, bias=False)

    def _radial_map(self, H, W, device):
        """Create normalised radial distance map [0,1] from image centre."""
        cy, cx = H / 2.0, W / 2.0
        ys = torch.arange(H, device=device).float()
        xs = torch.arange(W, device=device).float()
        yy, xx = torch.meshgrid(ys, xs, indexing='ij')
        dist = torch.sqrt(((yy - cy) / cy) ** 2 + ((xx - cx) / cx) ** 2)
        return dist.unsqueeze(0).unsqueeze(0)   # (1,1,H,W)

    def forward(self, x):
        B, C, H, W = x.shape
        radial = self._radial_map(H, W, x.device).expand(B, -1, -1, -1)
        # Gaussian-weighted radial map: peaks AT the rings, not just centre
        ring_bias = torch.exp(-((radial - 0.5) ** 2) / (2 * self.sigma ** 2 + 1e-6))
        x_aug = torch.cat([x, ring_bias], dim=1)   # concat as extra channel
        return self.mix(x_aug)


class RingSpecificAttention(nn.Module):
    """
    Ring-Specific Attention Module (RSAM).
    Combines:
      • Channel attention  (SE-style) — 'which channels carry ring info'
      • Circular spatial attention    — 'where are the concentric rings'
    Patent claim: the combination of channel + circular-positional
    attention specifically optimised for concentric target scoring.
    """
    def __init__(self, channels, reduction=8):
        super().__init__()

        # ── Channel Attention ────────────────
        self.ca_pool = nn.AdaptiveAvgPool2d(1)
        self.ca_fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(channels, channels // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channels // reduction, channels, bias=False),
            nn.Sigmoid(),
        )

        # ── Circular Spatial Attention ───────
        self.spatial_conv = nn.Conv2d(2, 1, kernel_size=7, padding=3, bias=False)
        self.spatial_act  = nn.Sigmoid()

        # ── Circular Positional Bias ─────────  ← novel component
        self.circular_bias = CircularPositionalBias(channels)

        # ── Output projection ────────────────
        self.proj = nn.Sequential(
            nn.Conv2d(channels, channels, 1, bias=False),
            nn.BatchNorm2d(channels),
            nn.SiLU(inplace=True),
        )

    def forward(self, x):
        identity = x

        # 1. Channel attention
        ca_w = self.ca_fc(self.ca_pool(x)).view(x.size(0), x.size(1), 1, 1)
        x = x * ca_w

        # 2. Spatial attention (CBAM-style but ring-aware)
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        sa_w = self.spatial_act(self.spatial_conv(torch.cat([avg_out, max_out], dim=1)))
        x = x * sa_w

        # 3. Circular positional bias injection
        x = self.circular_bias(x)

        # 4. Residual
        return self.proj(x) + identity

# Sanity check
_rsam = RingSpecificAttention(128)
_t = torch.zeros(2, 128, 40, 40)
_out = _rsam(_t)
print("Opt-3 RSAM output shape:", _out.shape, " (should match input)")
del _rsam, _t, _out
```

```python
def siou_loss(pred_boxes, target_boxes, eps=1e-7):
    """
    SIoU Loss implementation.
    Boxes in (x_centre, y_centre, w, h) format, normalised [0,1].

    Components:
      • Angle cost   — penalises angular deviation from GT direction
      • Distance cost — penalises centre offset
      • Shape cost   — penalises w/h mismatch
      • IoU term     — base intersection over union
    Returns scalar mean loss.
    """
    # Convert cx,cy,w,h → x1,y1,x2,y2
    def xywh2xyxy(b):
        x1 = b[..., 0] - b[..., 2] / 2
        y1 = b[..., 1] - b[..., 3] / 2
        x2 = b[..., 0] + b[..., 2] / 2
        y2 = b[..., 1] + b[..., 3] / 2
        return torch.stack([x1, y1, x2, y2], dim=-1)

    pb = xywh2xyxy(pred_boxes)
    tb = xywh2xyxy(target_boxes)

    inter_x1 = torch.max(pb[..., 0], tb[..., 0])
    inter_y1 = torch.max(pb[..., 1], tb[..., 1])
    inter_x2 = torch.min(pb[..., 2], tb[..., 2])
    inter_y2 = torch.min(pb[..., 3], tb[..., 3])

    inter_area = (inter_x2 - inter_x1).clamp(0) * (inter_y2 - inter_y1).clamp(0)
    pred_area  = (pb[..., 2] - pb[..., 0]) * (pb[..., 3] - pb[..., 1])
    gt_area    = (tb[..., 2] - tb[..., 0]) * (tb[..., 3] - tb[..., 1])
    union_area = pred_area + gt_area - inter_area + eps
    iou        = inter_area / union_area

    # Centre distance vector
    pc_x = (pb[..., 0] + pb[..., 2]) / 2
    pc_y = (pb[..., 1] + pb[..., 3]) / 2
    gc_x = (tb[..., 0] + tb[..., 2]) / 2
    gc_y = (tb[..., 1] + tb[..., 3]) / 2

    # Enclosing box
    enc_x1 = torch.min(pb[..., 0], tb[..., 0])
    enc_y1 = torch.min(pb[..., 1], tb[..., 1])
    enc_x2 = torch.max(pb[..., 2], tb[..., 2])
    enc_y2 = torch.max(pb[..., 3], tb[..., 3])
    enc_w  = (enc_x2 - enc_x1).clamp(eps)
    enc_h  = (enc_y2 - enc_y1).clamp(eps)

    # ── Angle cost ────────────────────────────
    sigma = torch.sqrt((gc_x - pc_x) ** 2 + (gc_y - pc_y) ** 2 + eps)
    sin_alpha = torch.abs(gc_y - pc_y) / sigma
    sin_beta  = torch.abs(gc_x - pc_x) / sigma
    threshold = (2 ** 0.5) / 2
    sin_alpha = torch.where(sin_alpha > threshold, sin_beta, sin_alpha)
    angle_cost = 1 - 2 * torch.sin(torch.arcsin(sin_alpha) - math.pi / 4) ** 2

    # ── Distance cost ─────────────────────────
    rho_x = ((gc_x - pc_x) / enc_w) ** 2
    rho_y = ((gc_y - pc_y) / enc_h) ** 2
    dist_cost = 2 - angle_cost - (1 - rho_x) * (1 - rho_y)

    # ── Shape cost ────────────────────────────
    pw = pred_boxes[..., 2].clamp(eps)
    ph = pred_boxes[..., 3].clamp(eps)
    gw = target_boxes[..., 2].clamp(eps)
    gh = target_boxes[..., 3].clamp(eps)
    omega_w = torch.abs(pw - gw) / torch.max(pw, gw)
    omega_h = torch.abs(ph - gh) / torch.max(ph, gh)
    shape_cost = (1 - torch.exp(-omega_w)) ** 2 + (1 - torch.exp(-omega_h)) ** 2

    siou = iou - (dist_cost + shape_cost) / 2
    return (1 - siou).mean()

# Quick numerical test
_p = torch.tensor([[0.5, 0.5, 0.2, 0.2]])
_g = torch.tensor([[0.52, 0.52, 0.2, 0.2]])
print(f"Opt-4 SIoU loss (near-perfect box): {siou_loss(_p, _g).item():.4f}  (expect ~0)")
del _p, _g
```

```python
class DetectionHead(nn.Module):
    """Lightweight YOLO-style detection head for one scale."""
    def __init__(self, in_channels, num_classes, num_anchors=3):
        super().__init__()
        self.num_classes = num_classes
        self.num_anchors = num_anchors
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, in_channels * 2, 3, padding=1, bias=False),
            nn.BatchNorm2d(in_channels * 2),
            nn.SiLU(inplace=True),
        )
        # Output: num_anchors × (5 + num_classes) per cell
        self.pred = nn.Conv2d(in_channels * 2,
                              num_anchors * (5 + num_classes), 1)

    def forward(self, x):
        return self.pred(self.conv(x))


class BullseyeDetector(nn.Module):
    """
    Full patent-grade detector:
    GhostNetV2 → BiFPN → RSAM × 3 → Detection Heads × 3
    """
    def __init__(self, num_classes=1, bifpn_channels=128,
                 bifpn_layers=3, num_anchors=3):
        super().__init__()

        # Opt 1: GhostNetV2 backbone
        self.backbone = GhostNetV2Backbone(pretrained=True)

        # Opt 2: BiFPN neck
        self.neck = BiFPNNeck(
            in_channels=self.backbone.out_channels,
            bifpn_channels=bifpn_channels,
            num_layers=bifpn_layers,
        )

        # Opt 3: Ring-Specific Attention on each scale
        self.rsam = nn.ModuleList([
            RingSpecificAttention(bifpn_channels) for _ in range(3)
        ])

        # Detection heads (P3=small, P4=medium, P5=large objects)
        self.heads = nn.ModuleList([
            DetectionHead(bifpn_channels, num_classes, num_anchors)
            for _ in range(3)
        ])

    def forward(self, x):
        # Backbone
        feats = self.backbone(x)          # [P3, P4, P5]
        # BiFPN neck
        feats = self.neck(feats)          # [P3', P4', P5']
        # RSAM on each scale
        feats = [rsam(f) for rsam, f in zip(self.rsam, feats)]
        # Detection heads
        preds = [head(f) for head, f in zip(self.heads, feats)]
        return preds   # list of 3 prediction tensors

# Instantiate and test full model
device = 'cuda' if torch.cuda.is_available() else 'cpu'
full_model = BullseyeDetector(num_classes=NUM_CLASSES).to(device)
dummy = torch.zeros(1, 3, 640, 640).to(device)
preds = full_model(dummy)
print("Full model outputs (3 scales):", [p.shape for p in preds])
total_params = sum(p.numel() for p in full_model.parameters()) / 1e6
print(f"Total parameters: {total_params:.2f}M  (GhostNetV2+BiFPN+RSAM)")
del dummy, preds
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from ultralytics import YOLO
from pathlib import Path

# ──────────────────────────────────────────────────────
# CONFIG
# ──────────────────────────────────────────────────────
best_weights = "/content/runs/detect/bullseye_yolov12_full/weights/best.pt"
inference_model = YOLO(best_weights)

# Run validation on TEST set and save plots
val_results = inference_model.val(
    data="/content/data.yaml",
    split="test",
    save=True,
    plots=True,
    conf=0.25,
    iou=0.6,
    project="/content/runs/detect",   # ← FIXED: explicit save location
    name="test_eval",                  # ← FIXED: fixed folder name
    exist_ok=True,                     # ← FIXED: no new numbered folders
)

# ← FIXED: hardcoded path instead of relying on val_results.save_dir
val_dir = Path("/content/runs/detect/test_eval")

# Debug — print exactly what files were saved
print("Save dir:", val_dir)
print("Files found:")
for f in sorted(val_dir.iterdir()):
    print(" ", f.name)

# ──────────────────────────────────────────────────────
# PLOT 1 — Core metric curves (PR, F1, P, R)
# ──────────────────────────────────────────────────────
curve_files = {
    "PR Curve"        : val_dir / "BoxPR_curve.png",   # ← was PR_curve.png
    "F1 Curve"        : val_dir / "BoxF1_curve.png",   # ← was F1_curve.png
    "Precision Curve" : val_dir / "BoxP_curve.png",    # ← was P_curve.png
    "Recall Curve"    : val_dir / "BoxR_curve.png",    # ← was R_curve.png
}

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Test Set — Metric Curves", fontsize=16, fontweight='bold')

for ax, (title, path) in zip(axes.flatten(), curve_files.items()):
    if path.exists():
        img = mpimg.imread(str(path))
        ax.imshow(img)
        ax.set_title(title, fontsize=13)
    else:
        ax.text(0.5, 0.5, f"{title}\nnot found\n{path}",   # ← FIXED: show full path in error
                ha='center', va='center', fontsize=9,
                transform=ax.transAxes)
    ax.axis("off")

plt.tight_layout()
plt.savefig("/content/test_metric_curves.png", dpi=150, bbox_inches='tight')
plt.show()
print("Saved → /content/test_metric_curves.png")

# ──────────────────────────────────────────────────────
# PLOT 2 — Confusion Matrix
# ──────────────────────────────────────────────────────
cm_paths = [
    val_dir / "confusion_matrix.png",
    val_dir / "confusion_matrix_normalized.png",
]

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle("Test Set — Confusion Matrix", fontsize=16, fontweight='bold')
titles = ["Confusion Matrix", "Confusion Matrix (Normalized)"]

for ax, path, title in zip(axes, cm_paths, titles):
    if path.exists():
        img = mpimg.imread(str(path))
        ax.imshow(img)
        ax.set_title(title, fontsize=13)
    else:
        ax.text(0.5, 0.5, f"{title}\nnot found\n{path}",   # ← FIXED: show full path
                ha='center', va='center', fontsize=9,
                transform=ax.transAxes)
    ax.axis("off")

plt.tight_layout()
plt.savefig("/content/test_confusion_matrix.png", dpi=150, bbox_inches='tight')
plt.show()
print("Saved → /content/test_confusion_matrix.png")

# ──────────────────────────────────────────────────────
# PLOT 3 — Sample predictions on test images
# ──────────────────────────────────────────────────────
test_img_dir = "/content/test/images"

# ← FIXED: handle missing folder gracefully
if not os.path.exists(test_img_dir):
    test_img_dir = "/content/valid/images"
    print(f"[warn] test/images not found, falling back to valid/images")

import random
all_images = sorted(os.listdir(test_img_dir))
test_images = random.sample(all_images, min(6, len(all_images)))

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle("Test Set — Sample Predictions", fontsize=16, fontweight='bold')

for ax, img_name in zip(axes.flatten(), test_images):
    img_path = os.path.join(test_img_dir, img_name)
    result   = inference_model.predict(img_path, conf=0.25, verbose=False)[0]
    annotated = result.plot()
    annotated = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)
    ax.imshow(annotated)
    ax.set_title(img_name[:30], fontsize=9)
    ax.axis("off")

for ax in axes.flatten()[len(test_images):]:
    ax.axis("off")

plt.tight_layout()
plt.savefig("/content/test_sample_predictions.png", dpi=150, bbox_inches='tight')
plt.show()
print("Saved → /content/test_sample_predictions.png")

# ──────────────────────────────────────────────────────
# PLOT 4 — Summary metrics bar chart
# ──────────────────────────────────────────────────────
metrics = {
    "mAP@0.5"      : val_results.box.map50,
    "mAP@0.5:0.95" : val_results.box.map,
    "Precision"    : val_results.box.mp,
    "Recall"       : val_results.box.mr,
}

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(metrics.keys(), metrics.values(),
              color=["#4C72B0","#DD8452","#55A868","#C44E52"],
              width=0.5, edgecolor='white', linewidth=1.2)

for bar, val in zip(bars, metrics.values()):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 0.01,
            f"{val:.4f}", ha='center', va='bottom', fontsize=12, fontweight='bold')

ax.set_ylim(0, 1.1)
ax.set_ylabel("Score", fontsize=12)
ax.set_title("Test Set — Patent-Grade Metrics Summary", fontsize=14, fontweight='bold')
ax.axhline(y=0.9, color='green', linestyle='--', alpha=0.6, label='Patent target (0.90)')
ax.legend(fontsize=10)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig("/content/test_metrics_summary.png", dpi=150, bbox_inches='tight')
plt.show()
print("Saved → /content/test_metrics_summary.png")

# ──────────────────────────────────────────────────────
# PRINT FINAL SUMMARY
# ──────────────────────────────────────────────────────
print("\n══ PATENT-GRADE TEST SET METRICS ══════════════")
print(f"  mAP@0.5       : {val_results.box.map50:.4f}")
print(f"  mAP@0.5:0.95  : {val_results.box.map:.4f}")
print(f"  Precision     : {val_results.box.mp:.4f}")
print(f"  Recall        : {val_results.box.mr:.4f}")
print(f"  Evaluated on  : TEST SET ✅")
print(f"  Results saved : {val_dir}")
print("═══════════════════════════════════════════════")
```
